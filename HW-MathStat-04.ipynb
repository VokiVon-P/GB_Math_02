#### Задача 1
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

#### Задача 2
О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.


D = (x2 - x1)**2/12 

12 * D = (x2 - x1)**2

sqrt(12*D) = x2-x1

x2 = sqrt(12*D)+x1

```python
x2 = np.sqrt(12 *0.2) + 0.5
print(f"Правая граница величины B = {x2}")
```

#### Задача 5
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

Нормируем данные для получения ответа.


```python
sigma = np.sqrt(25)
Z = (190 - 178)/sigma
print(f"Отклоняется на {Z} сигм")
```
